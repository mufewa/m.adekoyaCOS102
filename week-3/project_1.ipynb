```python
import csv
import tkinter as tk
from tkinter import messagebox as mb

rows = []

with open("GIG-logistics(in).csv", newline="") as file:
    reader = csv.reader(file)
    header = next(reader)

    for row in reader:
        rows.append(row[1:3])  # name, department


def enter():
    username = username_entry.get().strip()
    department = dep_entry.get().strip()

    if username and department:
        if [username, department] in rows:
            welcome(username, department)
        else:
            mb.showerror("Login Failed", "Invalid username or department.")
    else:
        mb.showwarning("Input Error", "Please enter both username and department.")


def welcome(username, department):
    mb.showinfo("Login Successful", f"Welcome, {username} from the {department} department")

    members = [name for name , dep in rows if dep == department and name != username]

    member_list = "\n".join(members)

    mb.showinfo(
        "Department Members",
        f"Other members in {department}:\n\n{member_list}"
    )


root = tk.Tk()
root.title("Login Page")
root.geometry("500x200")

username_label = tk.Label(root, text="Username:")
username_label.pack()

username_entry = tk.Entry(root)
username_entry.pack()

dep_label = tk.Label(root, text="Department:")
dep_label.pack()

dep_entry = tk.Entry(root)
dep_entry.pack()

submit_button = tk.Button(root, text="Enter", command=enter)
submit_button.pack()

root.mainloop()
```
