```python
import tkinter as tk
from tkinter import messagebox as msgbox
import pandas as pd

# Load the data
try:
    df = pd.read_csv('GIG-logistics(in).csv')
    # Clean up column names (removes extra spaces)
    df.columns = df.columns.str.strip() 
except FileNotFoundError:
    print("Error: GIG-logistics.csv not found!")

def verify_employee():
    user_name = name_entry.get().strip()
    user_dept = dept_entry.get().strip()

    # Search logic: Check if Name and Dept exist in the same row
    # Adjust 'NAME' and 'DEPARTMENT' to match the exact column names in your CSV
    match = df[(df['NAME'].str.lower() == user_name.lower()) & 
               (df['DEPARTMENT'].str.lower() == user_dept.lower())]

    if not match.empty:
        # Get coworkers (everyone in the same dept, excluding the user)
        coworkers = df[df['DEPARTMENT'].str.lower() == user_dept.lower()]
        coworker_list = coworkers['NAME'].tolist()
        
        # Show success and coworkers
        welcome_text = f"Welcome {user_name}!\n\nOther members in {user_dept}:\n"
        welcome_text += "\n".join(coworker_list)
        
        msgbox.showinfo("Access Granted", welcome_text)
    else:
        msgbox.showerror("Access Denied", "Employee does not exist or Department is incorrect.")

# --- GUI SETUP ---
root = tk.Tk()
root.title("GIG Logistics - Name Verification")
root.geometry("400x300")

tk.Label(root, text="Enter Name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department:").pack(pady=5)
dept_entry = tk.Entry(root)
dept_entry.pack()

tk.Button(root, text="Verify Identity", command=verify_employee).pack(pady=20)

root.mainloop()
```
