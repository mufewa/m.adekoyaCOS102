## Part A — Store Checkout After Purchase
### Pseudocode
    BEGIN StoreCheckout
      INPUT items[]
      SET total = 0
      FOR each item IN items DO
          total = total + item.price
      END FOR
      DISPLAY total
      INPUT paymentMethod
      IF paymentMethod == cash THEN
          INPUT amountGiven
          IF amountGiven >= total THEN
              change = amountGiven - total
              DISPLAY Payment successful. Change: change
          ELSE
              DISPLAY Insufficient funds. Please try again.
          END IF
      ELSE IF paymentMethod == card THEN
          PROCESS card payment
          IF payment approved THEN
              DISPLAY Payment successful.
          ELSE
              DISPLAY Card declined. Please try again.
          END IF
      END IF
      PRINT receipt
      DISPLAY Thank you for shopping with us!
    END StoreCheckout

## Part C — Swap Ages Algorithm
### Pseudocode
    BEGIN SwapAges
      INPUT name1
      INPUT age1
      INPUT name2
      INPUT age2
      DISPLAY Before swap:
      DISPLAY name1 -> age1
      DISPLAY name2 -> age2
      temp = age1
      age1 = age2
      age2 = temp
      DISPLAY After swap:
      DISPLAY name1 -> age1
      DISPLAY name2 -> age2
    END SwapAges

### Python Implementation

```python
def swap_ages():
    print('   AGE SWAPPER - Project I (C)')

    name1 = input('Enter first person name  : ')
    age1  = int(input('Enter ' + name1 + ' age : '))

    name2 = input('Enter second person name : ')
    age2  = int(input('Enter ' + name2 + ' age : '))

    print('Before Swap:')
    print('  ' + name1 + ': ' + str(age1))
    print('  ' + name2 + ': ' + str(age2))

    temp = age1
    age1 = age2
    age2 = temp

    print('After Swap:')
    print('  ' + name1 + ': ' + str(age1))
    print('  ' + name2 + ': ' + str(age2))
    print('Ages swapped successfully!')

swap_ages()
```
