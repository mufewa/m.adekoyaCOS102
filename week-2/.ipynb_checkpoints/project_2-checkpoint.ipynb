```python
import cmath
import math



def solve_quadratic():
    """Solve Ax² + Bx + C = 0 using the general quadratic formula:
       x = (-B ± √(B² - 4AC)) / 2A
    """
    print("\n── Quadratic Equation: Ax² + Bx + C = 0 ──")
    print("   Formula: x = (-B ± √(B² - 4AC)) / 2A")
    print()
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))

    if A == 0:
        print("Error: A cannot be 0 for a quadratic equation.")
        return

    discriminant = B**2 - 4*A*C


    root1 = (-B + cmath.sqrt(discriminant)) / (2 * A)   # x = (-B + √(B²-4AC)) / 2A
    root2 = (-B - cmath.sqrt(discriminant)) / (2 * A)   # x = (-B - √(B²-4AC)) / 2A

    print(f"\n  Discriminant (B² - 4AC) = {discriminant}")
    print(f"\n  x₁ = (-({B}) + √({discriminant})) / (2 × {A})")
    print(f"     = {format_complex(root1)}")
    print(f"\n  x₂ = (-({B}) - √({discriminant})) / (2 × {A})")
    print(f"     = {format_complex(root2)}")

    print()
    if discriminant > 0:
        print("→ Two distinct real roots")
    elif discriminant == 0:
        print("→ One repeated real root")
    else:
        print("→ Two complex (imaginary) roots")


def solve_cubic():
    """Solve Ax³ + Bx² + Cx + D = 0 using numpy-style numerical approach"""
    print("\n── Cubic Equation: Ax³ + Bx² + Cx + D = 0 ──")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))

    if A == 0:
        print("A cannot be 0 for a cubic equation. Solving as quadratic instead...")
        return


    a = B / A
    b = C / A
    c = D / A

    p = b - (a**2) / 3
    q = (2 * a**3) / 27 - (a * b) / 3 + c

    # Discriminant
    delta = -(4 * p**3 + 27 * q**2)

    print(f"\nDiscriminant: {round(delta, 6)}")

    # Use Cardano's formula
    D_card = (q/2)**2 + (p/3)**3

    u = cmath.exp(2j * cmath.pi / 3)  # primitive cube root of unity

    if D_card.real >= 0 and p == 0 and q == 0:
        t1 = t2 = t3 = complex(0)
    else:
        S = cmath.cbrt(-q/2 + cmath.sqrt(complex(D_card)))
        T = cmath.cbrt(-q/2 - cmath.sqrt(complex(D_card)))
        t1 = S + T
        t2 = -(S + T)/2 + 1j * (S - T) * math.sqrt(3)/2
        t3 = -(S + T)/2 - 1j * (S - T) * math.sqrt(3)/2

    shift = a / 3
    root1 = t1 - shift
    root2 = t2 - shift
    root3 = t3 - shift

    print(f"Root 1: {format_complex(root1)}")
    print(f"Root 2: {format_complex(root2)}")
    print(f"Root 3: {format_complex(root3)}")

    if delta > 0:
        print("→ Three distinct real roots")
    elif delta == 0:
        print("→ Repeated root (at least two roots are equal)")
    else:
        print("→ One real root and two complex conjugate roots")


def solve_quartic():
    """Solve Ax⁴ + Bx³ + Cx² + Dx + E = 0 using numpy roots"""
    print("\n── Quartic Equation: Ax⁴ + Bx³ + Cx² + Dx + E = 0 ──")
    A = float(input("Enter A: "))
    B = float(input("Enter B: "))
    C = float(input("Enter C: "))
    D = float(input("Enter D: "))
    E = float(input("Enter E: "))

    if A == 0:
        print("A cannot be 0 for a quartic equation.")
        return

    # Using companion matrix / numerical method via numpy
    try:
        import numpy as np
        coefficients = [A, B, C, D, E]
        roots = np.roots(coefficients)
        print(f"\nRoots of the quartic equation:")
        for i, root in enumerate(roots, 1):
            print(f"  Root {i}: {format_complex(root)}")
    except ImportError:
        print("\n[numpy not available — using basic numerical solver]")
        print("Please install numpy for full quartic support: pip install numpy")


def format_complex(z):
    """Format a complex number cleanly."""
    z = complex(z)
    real = round(z.real, 6)
    imag = round(z.imag, 6)
    if abs(imag) < 1e-9:
        return str(real)
    elif abs(real) < 1e-9:
        return f"{imag}i"
    else:
        sign = "+" if imag >= 0 else "-"
        return f"{real} {sign} {abs(imag)}i"


def main():
    print("=" * 50)
    print("       EQUATION ROOT FINDER — Project II")
    print("=" * 50)

    while True:
        print("\nSelect an operation:")
        print("  [1] Quadratic Equation  (Ax² + Bx + C = 0)")
        print("  [2] Cubic Equation      (Ax³ + Bx² + Cx + D = 0)")
        print("  [3] Quartic Equation    (Ax⁴ + Bx³ + Cx² + Dx + E = 0)")
        print("  [4] Exit")

        choice = input("\nEnter your choice (1–4): ").strip()

        if choice == "1":
            solve_quadratic()
        elif choice == "2":
            solve_cubic()
        elif choice == "3":
            solve_quartic()
        elif choice == "4":
            print("\nGoodbye!")
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 4.")

        again = input("\nSolve another equation? (y/n): ").strip().lower()
        if again != "y":
            print("\nGoodbye!")
            break


if __name__ == "__main__":
    main()
```
