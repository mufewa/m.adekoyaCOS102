```python
# Student data collection
students = [
    # Girls
    ("Evelyn", 17, 5.5, 80), ("Jessica", 16, 6.0, 85), ("Somto", 17, 5.4, 70),
    ("Edith", 18, 5.9, 60), ("Liza", 16, 5.6, 76), ("Madonna", 18, 5.5, 66),
    ("Waje", 17, 6.1, 87), ("Tola", 20, 6.0, 95), ("Aisha", 19, 5.7, 50),
    ("Latifa", 17, 5.5, 49),
    # Boys
    ("Chinedu", 19, 5.7, 74), ("Liam", 16, 5.9, 87), ("Wale", 18, 5.8, 75),
    ("Gbenga", 17, 6.1, 68), ("Abiola", 20, 5.9, 66), ("Kola", 19, 5.5, 78),
    ("Kunle", 16, 6.1, 87), ("George", 18, 5.4, 98), ("Thomas", 17, 5.8, 54),
    ("Wesley", 19, 5.7, 60)
]

def display_student_table(data):
    # Define table header
    header = f"{'NAME':<12} | {'AGE':<5} | {'HEIGHT':<8} | {'SCORE':<6}"
    separator = "-" * len(header)
    
    print("\nSTUDENT DATA TABLE")
    print(separator)
    print(header)
    print(separator)
    
    # Iterate through data and print formatted rows
    # <12 means left-aligned with 12 characters of space
    for name, age, height, score in data:
        print(f"{name:<12} | {age:<5} | {height:<8} | {score:<6}")
    
    print(separator)

# Execute the function
display_student_table(students)
```
