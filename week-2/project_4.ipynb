```python
def calculate_staff_atr():
    print("================================================")
    print("      IZIFIN TECHNOLOGY: ATR BATCH PROCESSOR    ")
    print("================================================\n")

    while True:
        try:
            # Taking inputs
            experience = int(input("Enter years of work experience: "))
            age = int(input("Enter age of the staff: "))
            
            # Logic to determine ATR
            if experience > 25 and age >= 55:
                atr = 5_600_000
            elif experience > 20 and age >= 45:
                atr = 4_480_000
            elif experience > 10 and age >= 35:
                atr = 1_500_000
            else:
                atr = 550_000
                
            # Display result
            print(f"RESULT: The ATR for this staff is N{atr:,}")
            print("-" * 30)

        except ValueError:
            print(">> Error: Please enter a whole number for age and experience.")
            continue

        # Ask the user if they want to continue or exit
        choice = input("Calculate for another staff? (yes/no): ").strip().lower()
        if choice not in ['yes', 'y']:
            print("\nExiting program. Have a productive day!")
            break

if __name__ == "__main__":
    calculate_staff_atr()
```
