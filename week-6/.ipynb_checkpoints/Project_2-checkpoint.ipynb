```python
import tkinter as tk
from tkinter import messagebox as mb
```

```python
class Delivery:
    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def calculate_fee(self):
        if self.location == "PAU" or self.location == "Pan-Atlantic University":
            self.location = "Pan-Atlantic University"
            if self.weight >= 10:
                return 2000
            else:
                return 1500

        elif self.location == "Epe" or self.location == "Lekki-Epe":
            self.location = "Lekki-Epe"
            if self.weight >= 10:
                return 5000
            else:
                return 4000

        else:
            return None
```

```python
def submit():
    location = location_entry.get().strip()

    try:
        weight = float(weight_entry.get())
        if weight < 0:
            mb.showerror("Error", "Weight cannot be negative.")
            return

    except ValueError:
        mb.showerror("Error", "Please enter a valid number for weight.")
        return

    delivery = Delivery(location, weight)
    fee = delivery.calculate_fee()

    if fee is None:
        mb.showerror("Invalid Location", "We only deliver to PAU or Epe.")
    else:
        weight = f"{weight:.2f}".rstrip("0").rstrip(".")
        mb.showinfo("Delivery Fee", f"Location: {location}\nWeight: {weight}kg\n\nAmount to pay: ₦{fee:,}")


root = tk.Tk()
root.title("Delivery Fee Calculator")
root.geometry("300x150")

# Location input
tk.Label(root, text="Enter Location (PAU or Epe):").pack()
location_entry = tk.Entry(root)
location_entry.pack()

# Weight input
tk.Label(root, text="Enter Package Weight (kg):").pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

# Submit button
tk.Button(root, text="Calculate Fee", command=submit).pack()

root.mainloop()
```
