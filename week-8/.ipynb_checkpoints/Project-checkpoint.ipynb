```python
import tkinter as tk
from tkinter import messagebox as mb
```

```python
class Zenith:
    def __init__(self, employee_name):
        self.employee_name = employee_name

    def unique_services(self):
        services = []

    def mutual_services(self):
        return ["Lines of credit",
            "Investments management and accounts",
            "Insurance"]
    
    def services(self):
        return self.mutual_services() + self.unique_services()

class RetailBanking(Zenith):
    def unique_services(self):
        return ["Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"]
    
class GlobalBanking(Zenith):
    def unique_services(self):
        return ["Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"]
    
    def mutual_services(self):
        return []
    
class CommercialBanking(Zenith):
    def unique_services(self):
        return ["Advisory services"]
```

```python
def on_enter():
    employee_name = name_entry.get()
    division = division_entry.get().lower()
    employee_map = {"Name": ["Mary", "Agatha", "Noel"],
                    "Division": ["Retail Banking", "Global Banking", "Commercial Banking"]}

    if employee_name not in employee_map["Name"]:
        if not employee_name:
            mb.showerror("Error", "Please enter a name.")
        else:
            mb.showerror("Error", "Invalid employee name. Please try again.")
        return
    elif division not in [d.lower() for d in employee_map["Division"]]:
        if not division:
            mb.showerror("Error", "Please enter a division.")
        else:
            mb.showerror("Error", "Invalid division. Please try again.")
        return

    for i in range(len(employee_map["Name"])):
        if employee_name == employee_map["Name"][i] and division == employee_map["Division"][i].lower():
            if division == "retail banking":
                employee = RetailBanking(employee_name)
            elif division == "global banking":
                employee = GlobalBanking(employee_name)
            elif division == "commercial banking":
                employee = CommercialBanking(employee_name)
            mb.showinfo("Services", f"Welcome, {employee_name}.\nServices rendered in your division:\n\n" + "\n".join(employee.services()))
            return
        
    mb.showerror("Error", "This person does not exist in this division. Please try again.")

root = tk.Tk()
root.title("Zenith Bank Employee Login")
root.geometry("400x200")

name_label = tk.Label(root, text="Enter your name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

division_label = tk.Label(root, text="Enter your division:")
division_label.pack()
division_entry = tk.Entry(root)
division_entry.pack()

enter = tk.Button(root, text="Enter", command=on_enter)
enter.pack()


root.mainloop()
```
